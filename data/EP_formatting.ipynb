# Format EmpatheticPersonas into DataFrames

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# I. Emotions Dataset 

```python
# choose language
lang = 'ZH' # 'ZH' or 'EN'

# import EmpatheticPersonas dataset
df = pd.read_csv('empatheticPersonas' + lang + '.csv', index_col=0)
df.head()
```

```python
# text  - extracts the 1181 expressions of emotions 
# label - emotion corresponding to text extract
if lang == 'ZH':
    emotions = ['悲伤', '愤怒', '快乐', '焦虑']
elif lang == 'EN':
    emotions = ['Sad', 'Angry', 'Happy', 'Anxious']

text = []
label = []
start_index = 0
label_encoding = 0
for emotion in emotions:
    for col in range(1,4):
        if lang == 'ZH':
            colname = emotion +' - 患者反应 ' + str(col)
        elif lang == 'EN':
            colname = emotion +' - Patient response ' + str(col)
        
        text += df[colname].dropna().values.tolist()
    end_index = len(text)
    # Label Encoding: 0 - 'Sad', 1- 'Angry', 2- 'Happy', 3 -'Anxious'
    label += [label_encoding]*(end_index - start_index)
    start_index = end_index
    label_encoding+=1

# print(text, label)
```

```python
# sanity check - check text and label same length
print(len(text))
print(len(label))
```

```python
# create the organised df
df = pd.DataFrame()
df['text']=text
df['labels']=label
```

# Check and Remove Overlap

```python
# function to check and remove overlap
def overlap_checker(df):
    df_list = df['text'].tolist()
    overlap_index = [index for (index,response) in enumerate(df_list) if response in df_list[index+1:]]
    
    return overlap_index
```

```python
# check for overlap
overlap = overlap_checker(df)
print(len(overlap)) # 21 instances in EN, 52 instances in ZH
```

```python
# remove overlap
df.drop(labels=overlap, axis=0, inplace=True)
```

```python
# Sanity Check - make sure no more overlaps
print(len(overlap_checker(df))) # 0
```

# Split Dataset

```python
# train-test-val split (80-10-10)
df_train, df_test = train_test_split(df, test_size=0.2, shuffle=True, random_state=0, stratify=df['labels'])
df_val, df_test = train_test_split(df_test, test_size=0.5, shuffle=True, random_state=0, stratify=df_test['labels'])
```

```python
# check distribution: approx equal
print(df_train['labels'].value_counts(normalize=True)*100)
print(df_test['labels'].value_counts(normalize=True)*100)
print(df_val['labels'].value_counts(normalize=True)*100)
```

```python
# export to csv
df.to_csv('emotions/EmpatheticPersonas/' + lang + '/emotionlabeled_full.csv', index=False)
df_train.to_csv('emotions/EmpatheticPersonas/' + lang + '/emotionlabeled_train.csv', index=False)
df_test.to_csv('emotions/EmpatheticPersonas/' + lang + '/emotionlabeled_test.csv', index=False)
df_val.to_csv('emotions/EmpatheticPersonas/' + lang + '/emotionlabeled_val.csv', index=False)
```

```python
# translate-train-all dataset 
# ZH_train + EN_full
zh_train = pd.read_csv('emotions/EmpatheticPersonas/ZH/emotionlabeled_train.csv', index_col=0)
en_full = pd.read_csv('emotions/EmpatheticPersonas/EN/emotionlabeled_train.csv', index_col=0)

train_all = pd.concat([zh_train, en_full])
train_all = train_all.sample(frac=1) # shuffle the dataset

train_all.to_csv('emotions/EmpatheticPersonas/EP_train.csv')
```

# II. Empathetic Rewritings

# General Formatting

```python
# choose language
lang = 'EN' # 'ZH' or 'EN'

# import EmpatheticPersonas dataset
df = pd.read_csv(f'empatheticPersonas{lang}.csv')
df.head()
```

```python
# Extracts the 2144 Empathetic Rewritings and format them into the following columns:
# sex, age, emotion, base, rewriting
df_formatted = pd.DataFrame(columns=['gender','age','emotion','base','rewriting'])

for column in df.columns[2:]:
    if '患者反应' not in column and 'Patient response' not in column:
        # get the rows where there is data in this column
        rows = df.index[df[column].isnull()==False].tolist()
        col_split = column.split(' - ') 
        if len(col_split) > 2:
            print ('ERROR: MORE THAN ONE SPLIT POINT')
            break
        emotion = [col_split[0]]*len(rows) 
        base = [col_split[1]]*len(rows)
        sex = df.iloc[rows,0]
        age = df.iloc[rows,1]
        rewrite = df[column].iloc[rows]
        df_loop = pd.DataFrame(list(zip(sex, age, emotion, base, rewrite)),
                                    columns =['gender','age','emotion','base','rewriting'])

        df_formatted = pd.concat([df_formatted, df_loop], ignore_index=True)
```

```python
# Replace age group 
new_age_group = []
for age in df_formatted['age']:
    if age in ['18-24','25-29','30-39']:
        new_age_group.append('18-39')
    else:
        new_age_group.append('40-69')

print(len(new_age_group))

df_formatted['age']=new_age_group
```

```python
# Export
df_formatted.to_csv(f'empathy/EP_empathy_2144_{lang}.csv')
```

# Formatting for Semantic Classifier

Labelling:</br>
1 - Was this caused by a specific event/s? | 这是由特别事件引起的吗？</br>
2 - Was this caused by a recent or distant event (or events)? | 这是由最近或遥远的事件（或多个事件）引起的吗？</br>
3 - Have you recently attempted protocol 6 and found this reignited unmanageable emotions as a result of old events? | 您最近是否尝试过协议6，并发现由于旧事件而重新点燃了无法控制的情绪？</br>
4 - Have you recently attempted protocol 11 and found this reignited unmanageable emotions as a result of old events? | 您最近是否尝试过协议11，并发现由于旧事件而重新点燃了无法控制的情绪？</br>
5 - Thank you. Now I will ask some questions to understand your situation.| 谢谢。现在我会问一些问题以了解您的情况。</br>
6 - Have you strongly felt or expressed any of the following emotions towards someone | 您是否对某人有强烈的感受或表达以下任何情绪：</br>
7 - Do you believe that you should be the saviour of someone else? | 您认为您应该成为别人的救星吗？</br>
8 - Do you see yourself as the victim, blaming someone else for how negative you feel? | 您是否将自己视为受害者，将自己的负面情绪归咎于他人？</br>
9 - Do you feel that you are trying to control someone? | 您觉得您在试图控制某人吗？</br>
10 - Are you always blaming and accusing yourself for when something goes wrong? | 当出现问题时，您是否总是责怪和指责自己？</br>
11 - In previous conversations, have you considered other viewpoints presented? | 在之前的对话中，您是否考虑过其他提出的观点？</br>
12 - Are you undergoing a personal crisis (experiencing difficulties with loved ones e.g. falling out with friends)? | 您是否正在经历个人危机（与亲人相处有困难，例如与朋友闹翻）？</br>
13 - That's Good! Let me recommend a protocol you can attempt. | 那很好！让我推荐一个您可以尝试的协议。</br>
14 - From what you have said I believe you are feeling {}. Is this correct? | 根据您所说的，我相信您正在感受{}。这个对吗？</br>
15 - I am sorry. Please select from the emotions below the one that best reflects what you are feeling: | 我很抱歉。请从下面的情绪中选择最能反映您感受的情绪：</br>
16 - Thank you for taking part. See you soon | 感谢您的参与。再见</br>
17 - Here are my recommendations, please select the protocol that you would like to attempt | 这是我的建议，请选择您想尝试的协议 </br>
18 - Please try to go through this protocol now. When you finish, press 'continue' | 请现在尝试通过此协议。完成后，按“继续”</br>
19 - Do you feel better or worse after having taken this protocol? | 采取此协议后，您感觉更好还是更糟？</br>
20 - Would you like to attempt another protocol? (Patient feels better) | 您想尝试另一种协议吗？ （病人感觉好多了）</br>
21 - Would you like to attempt another protocol? (Patient feels worse) | 您想尝试另一种协议吗？ （病人感觉更糟）</br>

```python
df_EN = pd.read_csv('empathy/EP_empathy_2144_EN.csv')
df_ZH = pd.read_csv('empathy/EP_empathy_2144_ZH.csv')

# Maps the base utterance to the numeric label (see above md text description)
def semantic_label(base_utterance_list):
    label = 1 
    label_list = []
    for idx in range(len(base_utterance_list)):
        if base_utterance_list[idx] in ['Was this caused by a specific event/s?', '这是由特别事件引起的吗？']:
            label = 1

        elif base_utterance_list[idx] != base_utterance_list[idx-1]:
            label += 1
        
        
        label_list.append(label)
    
    return label_list

label_list_EN = semantic_label(df_EN['base'].tolist())
label_list_ZH = semantic_label(df_ZH['base'].tolist())

# Sanity Check
print(label_list_EN==label_list_ZH)
```

```python
# Create a new df
label_list_concat = label_list_ZH + label_list_EN
rewritings_concat = pd.concat([df_ZH['rewriting'], df_EN['rewriting']], ignore_index=True)

df_semantic = pd.DataFrame(zip(rewritings_concat, label_list_concat), columns=['text','labels'])

df_semantic['labels'].value_counts()
```

```python
# Export

df_semantic.to_csv('empathy/semantic_labelled.csv')
```
